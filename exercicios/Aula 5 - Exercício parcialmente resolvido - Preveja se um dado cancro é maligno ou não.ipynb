## Problema

Neste exercício, pretendemos prever se um dado cancro é maligno ou não. Para tal, vamos ter em consideração diversas características dos núcleos celulares presentes em imagens de tumores, tais como a textura, o perímetro e a área do tumor.

## Dados

* Número de observações = 569
* Número de variáveis independentes = 31
 * Para informação detalhada sobre cada uma das variáveis, consultar: https://www.kaggle.com/uciml/breast-cancer-wisconsin-data
* Variável dependente = “diagnosis”. Diz-nos se o cancro é maligno (M) ou benigno (B).

Dados disponíveis em: https://raw.githubusercontent.com/pmarcelino/datasets/master/breast_cancer.csv

```
# Importar bibliotecas 
import pandas as pd  
import seaborn as sns

from sklearn.ensemble import RandomForestClassifier  
from sklearn.model_selection import cross_val_score
```

## 1. Preparar dados

```
# Importar dados

url = 'https://raw.githubusercontent.com/pmarcelino/datasets/master/breast_cancer.csv'
data = pd.read_csv(url)
data
```

```
# Verificar se há dados em falta

data.isnull().sum()
```

## 2. Explorar dados

```
# Visualizar a relação entre 'radius_mean' e 'area_mean', usando um scatterplot

sns.scatterplot(x="radius_mean", y="area_mean", data=data)
```

* Considerando as duas relações anteriores, consegues adivinhar a forma geométrica que um núcleo celular, aproximadamente, tem?

## 3. Treinar modelo

```
# Codificar a variável 'diagnosis'

data = pd.get_dummies(data,drop_first=True)
```

```
# Dividir em variáveis independentes e dependente
X = data[['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean', 
        'compactness_mean', 'concavity_mean','concave points_mean', 'symmetry_mean', 
        'fractal_dimension_mean']]
y = data['diagnosis_M']
```

```
# Definir algoritmo do modelo
model = RandomForestClassifier()
```

## 4. Avaliar modelo

```
# Avaliar modelo com validação cruzada (atenção à medida de avaliação a usar)

score = cross_val_score(model,X,y,scoring="recall",cv=5)
score
```

```
# Estimar desempenho do modelo

score_mean = score.mean()
score_mean
```
