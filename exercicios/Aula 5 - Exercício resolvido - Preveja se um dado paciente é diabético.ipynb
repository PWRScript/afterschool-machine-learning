# Exercício resolvido em branco

## Enunciado

Neste exercício, pretendemos prever se um dado paciente é diabético ou não. Para tal, vamos ter em consideração diversas características destes pacientes, tais como a sua idade, tensão arterial e índice de massa corporal.

## Dados

* Número de observações = 768
* Número de variáveis independentes = 9
 1. Pregnancies = número de gravidezes
 1. Glucose = nível de glicose (mg/dL)
 1. BloodPressure = pressão arterial diastólica (mmHg)
 1. SkinThickness = espessura da dobra da pele do tríceps (mm)
 1. Insulin = nível de insulina (mu U/ml)
 1. BMI = índice de massa corporal (kg/m^2)
 1. DiabetesPedigreeFunction = propensão genética para a diabetes (de acordo com uma função específica)
 1. Age = idade (anos)
* Número de variáveis dependentes = 1
 1. Outcome = diz-nos se um paciente é diabético (1) ou não (0)


Dados disponíveis em: https://raw.githubusercontent.com/pmarcelino/datasets/master/diabetes.csv

# Resolução

```
# importar bibliotecas
import pandas as pd
import seaborn as sb


from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```
# Obter dados
url = "https://raw.githubusercontent.com/pmarcelino/datasets/master/diabetes.csv"
data = pd.read_csv(url)
data
```

```
# Verificar dados em falta
data.isnull().sum()
```

```
# Verificar anormalias
data.describe()
```

```
# Corrigir anormalias
data = data.drop(data[data["Glucose"] == 0].index)
data = data.drop(data[data["Insulin"] == 0].index)
data = data.drop(data[data["BloodPressure"] == 0].index)
data = data.drop(data[data["SkinThickness"] == 0].index)
data = data.drop(data[data["BMI"] == 0].index)
```

```
# Verificar anormalias
data.describe()
```

```
# Definir variáveis independentes e dependente
X = data.drop("Outcome", axis=1)
y = data['Outcome']
```

```
# Definir algoritmo do modelo
model = RandomForestClassifier(random_state=20)
```

```
# Avaliar modelo validação cruzada (taxa acerto)
score = cross_val_score(model,X,y,cv=5)
score
```

```
# Estimar desempenho modelo
score_mean = score.mean()
score_mean
```

```
# Avaliar modelo validação cruzada (sensibilidade)
score = cross_val_score(model,X,y,scoring="recall",cv=5)
score
```

```
# Estimar desempenho modelo
score_mean = score.mean()
score_mean
```
